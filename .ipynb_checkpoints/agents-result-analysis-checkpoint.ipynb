### Defining variables

```python
# Getting model:generation(list) as a dict
models = os.listdir('models/')
models_gens = {}
for model in models:
    generations = os.listdir(os.path.join('models/', model))
    models_gens[model] = generations
```

```python
models
```

```python
def calculate_stats(model):
    results = {}
    
    def get_gen_number(gen):
        regex_number = re.compile(r'(\d+)') 
        return int(regex_number.findall(gen)[0])
    
    def open_file(filepath):
        with open(filepath, 'rb') as fp:
            population = pickle.load(fp)
        return population
    
    for gen in models_gens[model]:
        population = open_file(os.path.join('models/', model, gen))
        values = list(population.values())
        n_gen = get_gen_number(gen)
        results[n_gen] = [np.max(values), np.min(values), np.std(values), np.mean(values)]
        
    return results

results_j = calculate_stats('pop_50_gen_50__mr_02__1248_simplenet_victor_es_strategy_glorot_unif')
```

```python
# Renaming long models name to easier ones
model_to_experiment = {'pop_100_gen_100__mr_03__1248_simplenet_victor': 'F',
                     'pop_100_gen_100__mr_03__1578_simplenet_victor': 'H',
                     'pop_100_gen_100__mr_03__2346_simplenet_nh_20_victor': 'I',
                     'pop_50_gen_100__mr_03__13458_simplenet_nh_20_victor': 'J',
                     #'pop_50_gen_40__mr_02__1248_victor': 'A',
                     'pop_50_gen_50__mr_02__1248_simplenet_victor': 'C',
                     'pop_50_gen_50__mr_02__1248_simplenet_victor_es_strategy': 'E',
                     'pop_50_gen_50__mr_02__1248_simplenet_victor_es_strategy_glorot_unif': 'G',
                     'pop_50_gen_50__mr_02__1348_simplenet_victor': 'D',
                     #'pop_50_gen_50__mr_02__1348_victor': 'B'
                      }

models.remove('pop_50_gen_50__mr_03__1248_simplenet_victor')
models.remove('pop_50_gen_40__mr_02__1248_victor')
models.remove('pop_50_gen_50__mr_02__1348_victor')
```

```python
def building_dataframe(models):
    df_list = []
    
    def preprocess_df(df):
        df = df.reset_index()
        df.columns = ['generation', 'max_fitness', 'min_fitness', 'std_fitness', 'mean_fitness']
        df = df.sort_index()
        df['experiment'] = model_to_experiment[model]
        
        
        return df
    for model in models:
        print("Filling information for experiment {}...".format(model_to_experiment[model]))
        results = pd.DataFrame(calculate_stats(model)).T
        results = preprocess_df(results)
        
        df_list.append(results)
    
    final_df = pd.concat(df_list, axis=0)
    return final_df

df = building_dataframe(models)
```

### Mean Fitness for each experiment

```python
fig, ax = plt.subplots(4, 2, figsize=(10, 14))

def multiple_plots(df):
    for i, row in enumerate(df.experiment.unique().reshape(ax.shape)):
        for j, col in enumerate(row):
            df_slice = df[df.experiment==col]
            df_slice = df_slice.set_index('generation')
            sns.lineplot(
                         data=df_slice[['mean_fitness', 'min_fitness', 'max_fitness', 'std_fitness']],
                         ax=ax[i][j],
                           legend=False)
            ax[i][j].set_title("Experiment {}".format(col),
                              fontdict={'fontsize': '12'})
            ax[i][j].set_xlabel('')
            # plt.setp(ax[i][j].get_legend().get_texts(), fontsize='12')
multiple_plots(df)
#fig.legend(handles, labels, loc='upper right')
```

We can notice some interesting patterns about the combinations used in the parameters and the ones typical to genetic algorithms approaches. 
- First, for the green line representing the best agent in that specific generation, we have a ladder type of graphic, indicating that the solution got stucked in a local minimum before crossover and mutation happened to increase its value. 
- Second, we can see that we do not guarantee always that the best solution in one generation is passed to the following one. Since we did not use Elitist strategy, it sometimes happened that a solution with high fitness got discarded.  
- Third, also as expected, standard deviation for all experiments diminish as the generation number increases. Intuitively, this is narrowing our population to a final solution, we loose in exploitation and gain in exploration of a promise point.
- Even though neuroevolution is a probabilistic approach, the mean fitness increases in almost all tests.
- G and E are implemented using Evolution Strategy, that is, it implements a crossover based on real values. We can see that their fitness values are smoother than other experiments, probably due to this approach.
- H and I are the same set of parameters applied to different enemies. Of course, this is only one iteration, but we can see how choosing different enemies can have huge influence in our fitness values and how our algorithm performs regarding local minimum.

### Best of each experiment

Now, we will get the best solution for each experiment and test in all enemies to get the best one that will be our final agent.

**Note:** We are going to choose the best three solution for the last generation only.

```python
# Defining controller

from evoman.environment import Environment
experiment_to_last_gen = df.groupby('experiment').generation.count().to_dict()
def run_all_enemies(model):
    n = experiment_to_last_gen[model_to_experiment[model]]
    filepath = os.path.join('models', model, 'state_generation_'+ str(n-1) +'.pkl')
    with open(filepath, 'rb') as fp:
        population = pickle.load(fp)

    population_ordered = dict(sorted(population.items(), key=lambda x: x[1], reverse=True))
    i = 0
    results = {}
    for key, value in population_ordered.items():
        print(value)
        
        env = Environment(speed="fastest",
                          enemymode="static",
                          player_controller=key,
                          enemies=[1, 2, 3, 4, 5, 6, 7, 8],
                          multiplemode="yes",
                          level=2,
                         logs='off')

        env.update_parameter('contacthurt', 'player')

        f, p, e, t = env.play()
        results[i] = f
        
        
        if i==2:
            break
            
        i += 1 
    return results
```

```python
all_enemies_results = {}
for model in models:
    all_enemies_results[model_to_experiment[model]] = run_all_enemies(model)
```
